## üíé Advanced Customer Clustering with K-Means

## üß© Problem Statement
This project performs an in-depth analysis of customer segmentation. It focuses on finding hidden patterns in consumer spending habits and income levels using the **K-Means Clustering** algorithm.

---

## üìê Technical Implementation
- **Libraries**: `seaborn` for advanced visuals and `ydata-profiling` for deep data insights.
- **Optimal K**: Using the **WCSS** (Elbow Method) to ensure precise segmentation.
- **Clustering Features**: 2D clustering using Income vs. Spending Score.

---

## üìö Steps Performed
1. **Data Cleaning**: Checking for null values and data types.
2. **Profiling**: Generating a full report of the dataset.
3. **K-Selection**: Plotting WCSS to identify the "Elbow" point.
4. **Clustering**: Training the KMeans model with the optimal $K$.
5. **Centroid Mapping**: Highlighting the center point of each customer group.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
from ydata_profiling import ProfileReport
import seaborn as sns 
```

```python
df=pd.read_csv("Mall_Customers.csv")
```

```python
df
```

```python
x=df.iloc[:,3:]
```

```python
x
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,15):
    kmean=KMeans(n_clusters=i,init="k-means++",random_state=30)
    kmean.fit(x)
    wcss.append(kmean.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,15),wcss)
```

### this is called albo method

```python
# in the above there are show the 5 or 6 cluester are build so 
```

```python
kmean_1=KMeans(n_clusters=5,init='k-means++',random_state=30)
```

```python
kmean_1.fit_predict(x)
```

```python
label=kmean_1.fit_predict(x)
```

```python
x['cluster number ']=kmean_1.fit_predict(x)
```

```python
x
```

```python
# --- SHORT PLOT CODE ---
plt.figure(figsize=(7, 5))

# Plot Income vs. Score, colored by 'cluster number'
plt.scatter(x['Annual Income (k$)'], x['Spending Score (1-100)'], c=label, s=50, cmap='Set1') 

plt.xlabel('Income (k$)')
plt.ylabel('Spending Score')
plt.title('Clusters')
plt.colorbar(label='Cluster ID')
plt.show()
```

## predict the cluster

```python
kmean_1.predict([[33,44]])
```
