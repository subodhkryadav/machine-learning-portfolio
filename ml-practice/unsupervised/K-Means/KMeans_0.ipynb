## üõ†Ô∏è Workflow & Logic
1. **EDA**: Data distribution check using Profiling Report.
2. **Elbow Method**: Finding the optimal number of clusters ($K$) by plotting Within-Cluster Sum of Squares (WCSS).
3. **Model**: Implementing `KMeans` from Scikit-Learn.
4. **Visualization**: Scatter plot to show customer groups.

###### problem is perform the kmean problem with the above dataset and how many cluster are create?

```python
from sklearn.cluster import KMeans
```

```python
## how many cluester form in this dataset so find it
```

```python
wcss=[]
for i in range(1,15):
    kmean=KMeans(n_clusters=i,init="k-means++",random_state=30)
    kmean.fit(x)
    wcss.append(kmean.inertia_)
```

```python
plt.plot(range(1,15),wcss)
plt.title("albo method")
```

```python
## here the k=5 or 6
```

```python
kmean_1=KMeans(n_clusters=5,init="k-means++",random_state=30)
```

```python
kmean_1.fit_predict(x)
```

```python
x["cluester number"]=kmean_1.fit_predict(x)
```
