## üìç DBSCAN (Density-Based Spatial Clustering of Applications with Noise)

### üß© Problem Statement
The goal is to perform **Customer Segmentation** using the Mall Customers dataset. Unlike K-Means, we use **DBSCAN** to identify clusters based on the density of data points and effectively handle outliers (noise).

---

### üìä Dataset Features
- **Annual Income (k$)**: Customer's yearly income.
- **Spending Score (1-100)**: Score assigned by the mall based on customer behavior.

---

### üõ†Ô∏è Key Steps Performed
1. **EDA**: Analyzing the distribution of income and spending.
2. **Feature Selection**: Using relevant numerical features for clustering.
3. **Model Implementation**: Applying the DBSCAN algorithm from `sklearn.cluster`.
4. **Parameter Tuning**: Setting `eps` (epsilon) and `min_samples` for optimal cluster formation.
5. **Evaluation**: Calculating the **Silhouette Score** for cluster quality.

---

### üìà Results & Visuals
- Identifying high-density clusters of shoppers.
- Detecting noise points (outliers) that do not belong to any specific group.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
from sklearn.cluster import DBSCAN
```

## load the dataset

```python
df=pd.read_csv("Mall_Customers.csv")
```

```python
df
```

```python
x=df.iloc[:,3:]
```

```python
x
```

```python
dbscan=DBSCAN(eps=1,min_samples=3)
```

```python
dbscan.fit(x)
```

```python
dbscan.labels_
```

## len of the the cluster

```python
set(dbscan.labels_)
```

```python
len(dbscan.labels_)
```

```python
labels=dbscan.labels_
```

```python
labels
```

```python
plt.figure(figsize=(7,7))
plt.scatter(x["Annual Income (k$)"],x["Spending Score (1-100)"],c=labels,s=50,cmap="Set1")
plt.xlabel('Income (k$)')
plt.ylabel('Spending Score')
plt.title('Clusters')
plt.colorbar(label='Cluster ID')
plt.show()
```

```python
from sklearn.metrics import silhouette_score

non_noise_mask = (labels != -1)
x_filtered = x[non_noise_mask]
labels_filtered = labels[non_noise_mask]

if len(np.unique(labels_filtered)) >= 2:
    score = silhouette_score(x_filtered, labels_filtered)
    print(f"{score:.4f}")
else:
    print("Score not computable.")
```
